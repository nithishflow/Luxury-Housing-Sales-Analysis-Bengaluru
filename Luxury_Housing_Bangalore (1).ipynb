```python
df['Ticket_Price_Cr']=df['Ticket_Price_Cr'].astype(str).str.replace('cr','',case=False,regex=False).str.replace('₹','').str.strip()
df['Ticket_Price_Cr'] = pd.to_numeric(df['Ticket_Price_Cr'], errors='coerce')

df.loc[df['Ticket_Price_Cr'] > 1e7, 'Ticket_Price_Cr'] /= 1e7
```

```python
df['Micro_Market']=df['Micro_Market'].astype(str).str.title().str.strip()
```

```python
df['Unit_Size_Sqft']=df['Unit_Size_Sqft'].fillna(df['Unit_Size_Sqft'].median())
```

```python
df['Amenity_Score']=df['Amenity_Score'].fillna(df['Amenity_Score'].median())
```

```python
df['Buyer_Comments']=df['Buyer_Comments'].fillna('no command')
```

```python
#This fills missing ticket prices using the median price of the same configuration type
df['Ticket_Price_Cr']=df.groupby('Configuration')['Ticket_Price_Cr'].transform(lambda x:x.fillna(x.median()))
```

```python
df.rename(columns={'Developer_Name':'Builder'},inplace=True)
```

```python
df['Builder']=df['Builder'].astype(str).str.title().str.strip()
```

```python
df['Configuration']=df['Configuration'].astype(str).str.lower()
```

```python
#This converts total price from crores to rupees and divides by unit size to calculate price per square foot
df['Price_per_Sqft']=(df['Ticket_Price_Cr']*1e7)/df['Unit_Size_Sqft']
```

```python
#This creates a binary flag where primary transactions are marked as booked (1) and all others as not booked (0)
df['Booking_Flag'] = np.where(df['Transaction_Type'].str.lower() == 'primary',1,0)
```

```python
#This converts purchase dates into proper datetime format and extracts the quarter number from each date
df['Purchase_Quarter'] = pd.to_datetime(df['Purchase_Quarter'])
df['Quarter_Number'] = df['Purchase_Quarter'].dt.quarter
```

```python
#This standardizes Yes/No values and converts them into a binary numeric flag
df['NRI_Buyer'] = (df['NRI_Buyer'].astype(str).str.lower().map({'yes': 1, 'no': 0}))
```
